##DATA CLEANING

###Categorical variables

```
# Inspect categorical variables
qualitative = []
for col_name in df.columns:
  if df[col_name].dtypes == 'object':
    unique_cat = len(df[col_name].unique())
    qualitative.append(col_name)
    print('Variable ' + str(col_name) + ' have ' + str(unique_cat) + ' categories ')
print('List of qualitative variables: '+ str(qualitative))
```

```
#Function get dummies(Data frame, categorical variables): Convert categorical variable into dummy/indicator variables. 
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```
#call dummy_df function
df = dummy_df(df, qualitative)
df.head()
```

```
#Inspect: no losing rows
df.shape
```

###Missing Values

```
#Fill null values with mean
df.fillna(df.mean(), inplace=True)
```
